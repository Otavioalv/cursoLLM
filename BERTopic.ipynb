```python
# pre processamento de dados
import glob

dados = []
dir = "/content/documentos/"

for arquivo in glob.iglob(dir + '*.txt', recursive=False):
  meuarquivo = open(arquivo, 'r', encoding='utf8')
  content = meuarquivo.read()
  dados.append(content)
  meuarquivo.close()


```

```python
len(dados)
```

```python
dados[3]
```

```python
# transformar a lista de documentos em uma  lista de sentenças
sentencas = []
for item in dados:
  for t in sent_tokenize(item):
    sentencas.append(t)

print(len(sentencas))

```

```python
sentencas[300]
```

```python
modelo = BERTopic(language="portuguese", verbose=True, top_n_words=15,
                  min_topic_size=10, nr_topics=20,
                  vectorizer_model=CountVectorizer(
                      ngram_range=(1, 3),
                      stop_words=stopwords.words('portuguese'),
                      min_df=10
                      )
                  )
```

```python
%%time
topicos, probabilidades = modelo.fit_transform(sentencas)
```
