```python
import transformers
from transformers import pipeline
```

```python
gerador = pipeline('text-generation', model="pierreguillou/gpt2-small-portuguese")
```

```python
texto = "O Brasil é um país tropical"

resultado = gerador(texto, max_length=20, do_sample=True)

print(resultado)

for r in resultado:
  print(r)
```
